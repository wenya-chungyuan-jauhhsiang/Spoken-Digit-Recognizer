## 處理流程
- 使用者選取Model，共六種Model供選擇，分別是
    - EN_CQT_CNN.h5
    - CH_CQT_CNN.h5
    - EN_STFT_CNN.h5
    - CH_STFT_CNN.h5
    - EN_MFCC_RNN.h5
    - CH_MFCC_RNN.h5
- 點選"Record"按鈕錄音，再按一次停止錄音
- 程式將使用split_audio.py之程式分割音檔
- 使用選擇的Model偵測每個分割的音檔數字為何
